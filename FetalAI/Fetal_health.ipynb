## Splitting into Train and Test Dataset

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
X_train.shape, X_test.shape
```

##Solving Class Imbalance

```python
smote=SMOTE()
smote
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train.astype('float'),y_train)
```

```python
from collections import Counter
print("Before SMOTE:",Counter(y_train))
print("After SMOTE:", Counter(y_train_smote))
```

##Training On Different Algorithms

###Random Forest

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train_smote, y_train_smote)
predictions=rf_model.predict(X_test)
print(accuracy_score(y_test,predictions))
```

```python
print("Accuracy of Random Forest Classifier: ",rf_model.score(X_test,y_test))
cm = confusion_matrix(y_test, predictions)
cm_display = ConfusionMatrixDisplay(cm).plot()
plt.show()
```

###Decision Tree

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train_smote, y_train_smote)
predictions = dt_model.predict(X_test)
print(accuracy_score(y_test,predictions))
```

```python
print("Accuracy of Decision Tree Classifier: ",dt_model.score(X_test,y_test))
cm = confusion_matrix(y_test, predictions)
cm_display = ConfusionMatrixDisplay(cm).plot()
plt.show()
```

###Logistic Regression

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_smote, y_train_smote)
predictions = lr_model.predict(X_test)
print(accuracy_score(y_test,predictions))
```

```python
print("Accuracy of Logistic Regression: ",lr_model.score(X_test,y_test))
cm = confusion_matrix(y_test, predictions)
cm_display = ConfusionMatrixDisplay(cm).plot()
plt.show()
```

###K-Nearest Neighbors

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_smote, y_train_smote)
predictions = knn_model.predict(X_test)
print(accuracy_score(y_test,predictions))
```

```python
print("Accuracy of K nearest Neighbors: ",knn_model.score(X_test,y_test))
cm = confusion_matrix(y_test, predictions)
cm_display = ConfusionMatrixDisplay(cm).plot()
plt.show()
```
